### Zepto

```python
# Zepto 
Z_orders = pd.read_csv("../data/raw/Zepto/zepto_order.csv")
Z_product = pd.read_csv("../data/raw/Zepto/zepto_product.csv")
Z_delivery= pd.read_csv("../data/raw/Zepto/zepto_delivery.csv")
Z_customer = pd.read_csv("../data/raw/Zepto/zepto_customer.csv")
Z_rating = pd.read_csv("../data/raw/Zepto/zepto_rating.csv")
Z_transaction= pd.read_csv("../data/raw/Zepto/zepto_transaction.csv")
```

```python

# Orders Table of Zepto
Z_orders.info()
print(4*"______________________")
print(Z_orders.shape)

print(4*"______________________")
print(Z_orders.isnull().sum())

print(4*"______________________")
print(Z_orders.nunique())

print(4*"______________________")
print(Z_orders.head())

print(4*"______________________")
print(Z_orders["order_status"].unique())
```

```python

# Product Table of Zepto
Z_product.info()
print(4*"______________________")
print(Z_product.shape)

print(4*"______________________")
print(Z_product.isnull().sum())

print(4*"______________________")
print(Z_product.nunique())

print(4*"______________________")
print(Z_product.head())

#Check product_name, sub_category, category name standardization
print(4*"______________________")
print(Z_product["product_name"].unique())

print(4*"______________________")
print(Z_product["sub_category"].unique())

print(4*"______________________")
print(Z_product["category"].unique())
```

```python
# Delivery Table of Zepto
Z_delivery.info()
print(4*"______________________")
print(Z_delivery.shape)

print(4*"______________________")
print(Z_delivery.isnull().sum())

print(4*"______________________")
print(Z_delivery.nunique())

print(4*"______________________")
print(Z_delivery.head())

print(4*"______________________")
print(Z_delivery["delivery_status"].unique())

# missing order id check between orders and delivery table
print(4*"______________________")
all_present = orders1['order_id'].isin(Z_delivery['order_id']).all()

if all_present:
    print("✅ Success! All  order_ids from the order table are present in the delivery table.")
else:
    print("❌ Match Failed! Some order_ids are missing from the delivery table.")
    missing_in_delivery = orders1[~orders1['order_id'].isin(Z_delivery['order_id'])]['order_id']
    print(f"Total missing orders: {len(missing_in_delivery)}")
    # print("First few missing IDs:\n", missing_in_delivery.head())
```

```python

# Customer Table of Zepto
Z_customer.info()
print(4*"______________________")
print(Z_customer.shape)

print(4*"______________________")
print(Z_customer.isnull().sum())

print(4*"______________________")
print(Z_customer.nunique())

print(4*"______________________")
print(Z_customer.head())

#check gender, city , state standardization 
print(4*"______________________")
print(Z_customer["gender"].unique())

print(4*"______________________")
print(Z_customer["city"].unique())

print(4*"______________________")
print(Z_customer["state"].unique())
```

```python

# Rating Table of Zepto
Z_rating.info()
print(4*"______________________")
print(Z_rating.shape)

print(4*"______________________")
print(Z_rating.isnull().sum())

print(4*"______________________")
print(Z_rating.nunique())

print(4*"______________________")
print(Z_rating.head())

#check review standardization
print(4*"______________________")
print(Z_rating["review"].unique())

#check all order_ids from orders table are present in rating table
print(4*"______________________")
all_present = orders1['order_id'].isin(Z_rating['order_id']).all()

if all_present:
    print("✅ Success! All order_ids from the order table are present in the rating table.")
else:
    print("❌ Match Failed! Some order_ids are missing from the rating table.")
    
    missing_in_rating =set(orders1[~orders1['order_id'].isin(Z_rating['order_id'])]['order_id'])
    print(f"Total missing orders: {len(missing_in_rating)}")
    # print("First few missing IDs:\n", missing_in_rating.head())
  
    
#check all order_ids from delivery table are present in rating table
print(4*"______________________")
all_present = Z_delivery['order_id'].isin(Z_rating['order_id']).all()

if all_present:
    print("✅ Success! All order_ids from the order table are present in the rating table.")
else:
    print("❌ Match Failed! Some order_ids are missing from the rating table.")
    delivery_missing_in_rating = Z_delivery[~Z_delivery['order_id'].isin(Z_rating['order_id'])]['order_id']
    print(f"Total missing orders: {len(delivery_missing_in_rating)}")
    # print("First few missing IDs:\n", delivery_missing_in_rating.head())
```

```python

# Transaction Table of Zepto
Z_transaction.info()
print(4*"______________________")
print(Z_transaction.shape)

print(4*"______________________")
print(Z_transaction.isnull().sum())

print(4*"______________________")
print(Z_transaction.nunique())

print(4*"______________________")
Z_transaction.head()

print(4*"______________________")
print(Z_transaction["payment_mode"].unique())

#check all order_ids from orders table are present in transaction table
print(4*"______________________")
all_present = orders1['order_id'].isin(Z_transaction['order_id']).all()

if all_present:
    print("✅ Success! All order_ids from the order table are present in the transaction table.")
else:
    print("❌ Match Failed! Some order_ids are missing from the transaction table.")
    missing_in_transaction = set(orders1[~orders1['order_id'].isin(Z_transaction['order_id'])]['order_id'])
    print(f"Total missing orders: {len(missing_in_transaction)}")
    # print("First few missing IDs:\n", missing_in_transaction.head())

# check missing order id are same or not between rating and transaction
print(4*"______________________")
if missing_in_rating == missing_in_transaction:
    print("✅ both table have same missing order ids")
else:
    print("❌both table have different missing order ids or something else.")
```
